```python
# Data wrangling
```

```python
import pandas as pd
import numpy as np
```

```python
SalesDf = pd.read_csv('Sales_August_2019.csv')
```

```python
# see the first 5
```

```python
SalesDf.head()
```

```python
SalesDf.info()
```

```python
# convert to int
```

```python
SalesDf['Quantity Ordered'] = pd.to_numeric(SalesDf['Quantity Ordered'], errors = 'coerce')
SalesDf
```

```python
SalesDf['Price Each'] = SalesDf['Price Each'].apply(pd.to_numeric, errors = 'coerce')
SalesDf
```

```python
SalesDf['Order Date'] = pd.to_datetime(SalesDf['Order Date'], errors = 'coerce')
SalesDf
```

```python
#check for missing values

SalesDf.isnull().sum()
```

```python
SalesDf.info()
```

```python
SalesDf.dropna(inplace = True)
SalesDf.isnull().sum()
```

```python
SalesDf.shape
```

```python
SalesDf.tail()
```

```python
# now this index is not matching the number of row, let's correct that
```

```python
SalesDf.reset_index(drop = True, inplace = True) # drop ensures the old index is droped so we do not have 2 indices
SalesDf.tail()
```

```python
#now let's plot graph using seaborn and matplotlib
# firt we import our libraries
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.barplot(data = SalesDf, x = 'Product', y = 'Quantity Ordered')
plt.title('Bar Chart of Quantity Ordered')
plt.xticks(rotation = 45, ha = 'right' )
plt.show()
```

```python
SalesDf['Quantity Ordered'].nunique()
```

```python
#let's find the total sales for each product so  we can know the product that sold the most
```

```python
SalesDf.columns
```

```python
SalesDf['Total Sales'] = SalesDf['Quantity Ordered'] * SalesDf['Price Each']
SalesDf.head()
```

```python
SalesDf['Total Sales'].max()
```

```python
# lets tabulate it so we can know the product
```

```python
SalesDf_BestSeller = SalesDf.groupby('Product')['Total Sales'].sum().reset_index()
SalesDf_BestSeller = SalesDf_BestSeller.sort_values(by='Total Sales', ascending=False)

SalesDf_BestSeller = SalesDf_BestSeller.reset_index(drop= True)
SalesDf_BestSeller
```

```python
plt.figure(figsize=(14, 6))  # Increase figure size for more space
sns.barplot(data = SalesDf_BestSeller, x = 'Product', y = 'Total Sales', width = 0.6, palette = 'tab10')
plt.title('Bar Chart of Total Sales')
plt.xticks(rotation=45, ha='right')  # Rotate labels diagonally for better readability

plt.tight_layout()  # Adjust layout to prevent labels from being cut off
plt.show()
```
